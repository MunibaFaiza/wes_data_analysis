```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating sample data
data = pd.DataFrame({
    "Chromosome": ["1", "1", "2", "2", "3", "3", "4", "4"],
    "Position": [10, 50, 20, 80, 30, 90, 40, 100],
    "G-Score": [1.5, -1.2, 2.3, -2.0, 1.8, -1.5, 2.0, -1.8]
})

print(data.head)
```

```python
# Plotting the G-score
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

colors = sns.color_palette("coolwarm", as_cmap=True)

# Creating a scatter plot (Gains are shown in red & Losses in blue)
sc = plt.scatter(
    data["Position"], data["G-Score"], 
    c=data["G-Score"], cmap=colors, s=200, edgecolors="black", alpha=0.8
)

cbar = plt.colorbar(sc)
cbar.set_label("G-Score Intensity")

plt.axhline(0, color="black", linestyle="--", linewidth=1)  # Baseline at 0
plt.xlabel("Genomic Position", fontsize=12)
plt.ylabel("G-Score", fontsize=12)
plt.title("CNV G-Score Representation (Gains & Losses)", fontsize=14, fontweight="bold")

plt.show()
```
