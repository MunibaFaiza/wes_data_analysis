```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the sample data
data = pd.read_csv("sample_data/bubble_plot_sample_data.csv")
print(data.head())
```

```python
plt.figure(figsize=(14, 7))

# Scatter plot
ax = sns.scatterplot(
    x=data["Gene"],
    y=data["Impact"],
    size=data["Impact"],
    hue=data["Chromosome"],
    palette="coolwarm",
    sizes=(50, 800),
    alpha=0.75,
    edgecolors="black",
    linewidth=1
)

plt.xlabel("Genes", fontsize=14, fontweight="bold", labelpad=10)
plt.ylabel("Impact Score", fontsize=14, fontweight="bold", labelpad=10)
plt.title("Genetic Alteration Impact - Bubble Plot", fontsize=16, fontweight="bold", pad=15)

plt.xticks(rotation=45, ha="right", fontsize=12)

plt.yticks(fontsize=12)

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.legend(title="Chromosome", bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()
```

# Bubble plot with Alteration types as colors

```python
# Loading the sample data
data = pd.read_csv("sample_data/bubble_plot_sample_data.csv")
print(data.head())
```

```python
# Creating the bubble plot with Alteration types as colors

plt.figure(figsize=(14, 7))

ax = sns.scatterplot(
    x=data["Chromosome"],
    y=data["Impact"],
    size=data["Impact"],
    hue=data["Alteration"],
    palette="Set2",
    sizes=(50, 800),
    alpha=0.8,
    edgecolors="black",
    linewidth=1
)

# Adding Gene names on bubbles
for i in range(len(data)):
    plt.text(
        x=data["Chromosome"][i], 
        y=data["Impact"][i], 
        s=data["Gene"][i], 
        fontsize=10, ha="center", va="center", 
        color="black", fontweight="bold"
    )

plt.xlabel("Chromosome", fontsize=14, fontweight="bold", labelpad=10)
plt.ylabel("Impact Score", fontsize=14, fontweight="bold", labelpad=10)
plt.title("Genetic Alteration Impact - Bubble Plot", fontsize=16, fontweight="bold", pad=15)

plt.xticks(rotation=45, ha="right", fontsize=12)

plt.yticks(fontsize=12)

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.legend(title="Alteration Type", bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()
```
