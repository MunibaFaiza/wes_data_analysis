## using sample data file

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the sample data file
data = pd.read_csv("sample_data/arm_level_correlation.csv")
print(data.head())

```

```python
plt.figure(figsize=(10, 6))

# Creating a heatmap
sns.heatmap(data.set_index("Chromosome_Arm"), cmap="Pastel1", annot=True, fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Correlation'})

plt.title("Correlation of Frequency of Arm-Level CNVs", fontsize=14, fontweight='bold')
plt.xlabel("Chromosomal Arms", fontsize=12)
plt.ylabel("Chromosome", fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

## Using random sample data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating random sample data
np.random.seed(42)
cancer_types = ["Lung", "Breast", "Colon", "Prostate", "Ovarian", "Leukemia", "Liver", "Pancreatic"]
chromosomal_arms = ["1p", "1q", "2p", "2q", "3p", "3q", "4p", "4q"]

data = pd.DataFrame(np.random.rand(len(chromosomal_arms), len(cancer_types)), 
                    columns=cancer_types, index=chromosomal_arms)

np.fill_diagonal(data.values, 1)

print(data.head())


```

```python
# Creating the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, fmt=".2f", cmap="Set2", linewidths=0.5, square=True, cbar_kws={"shrink": 0.8})

plt.title("Correlation of Frequency of Arm-Level Alterations", fontsize=14, fontweight='bold')
plt.xlabel("Cancer Type", fontsize=12)
plt.ylabel("Chromosomal Arm", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)


plt.show()
```
