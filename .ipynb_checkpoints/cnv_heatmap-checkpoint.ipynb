```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the sample data
cnv_data = pd.read_csv("sample_data/cnv_sample_data.csv")
print(cnv_data.head)
```

```python
# Creating the heatmap
cnv_data.set_index("Sample", inplace=True)

plt.figure(figsize=(12, 6))
sns.heatmap(cnv_data, cmap="coolwarm", annot=True, fmt=".1f", linewidths=0.5, cbar_kws={"label": "CNV Value"})

plt.title("Copy Number Variation (CNV) Heatmap", fontsize=14, fontweight="bold")
plt.xlabel("Chromosomes", fontsize=12)
plt.ylabel("Samples", fontsize=12)

plt.show()
```
