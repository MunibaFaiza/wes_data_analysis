```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn3
```

```python
# Loading the sample data
file_path = "sample_data/sample_wes_data.csv"  
wes_df = pd.read_csv(file_path)
print (wes_df.head())
```

```python
# Sorting data by chromosome and start position
wes_df.sort_values(by=["chromosome", "start"], inplace=True)
```

```python
# Selecting three representative chromosomes for the Venn diagram (You can select any chromosomes)
chrom1, chrom2, chrom3 = "chr1", "chr2", "chr3"
set1 = set(wes_df[wes_df["chromosome"] == chrom1]["gene"])
set2 = set(wes_df[wes_df["chromosome"] == chrom2]["gene"])
set3 = set(wes_df[wes_df["chromosome"] == chrom3]["gene"])
```

```python
# Plotting the Venn diagram
plt.figure(figsize=(6, 6))
venn3([set1, set2, set3], set_labels=(chrom1, chrom2, chrom3))
plt.title("Venn Diagram of CNVs in Different Chromosomes")
plt.show()
```

## Histogram of Copy Number Variants

```python
plt.figure(figsize=(8, 5))
sns.histplot(wes_df["variant_count"], bins=30, kde=True, color="royalblue")
plt.xlabel("Number of Variants (CNVs)")
plt.ylabel("Frequency")
plt.title("Histogram of Copy Number Variants")
plt.show()
```

## Histogram of GC Content 

```python
plt.figure(figsize=(8, 5))
sns.histplot(wes_df["GC_content"], bins=30, kde=True, color="green")
plt.xlabel("GC Content (%)")
plt.ylabel("Frequency")
plt.title("Histogram of GC Content Across Exons")
plt.show()
```
