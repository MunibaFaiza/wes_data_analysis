```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the sample whole exome data
file_path = "sample_data/sample_wes_data.csv"  # Update path if needed
wes_df = pd.read_csv(file_path)
print(wes_df.head())
```

```python
# Sorting data by chromosome and start position
wes_df.sort_values(by=["chromosome", "start"], inplace=True)
```

```python
# Setting plot style
sns.set(style="whitegrid")
```

```python
# Creating scatter plot
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    x = wes_df["start"], 
    y = wes_df["coverage"], 
    hue = wes_df["chromosome"], 
    palette = "tab20", 
    alpha = 0.8, 
    edgecolor = None
)

plt.xlabel("Genomic Position (Start Site)", fontsize=12)
plt.ylabel("Read Depth (Coverage)", fontsize=12)
plt.title("Scatter Plot of Read Depth Across Exons", fontsize=14)
plt.legend(title="Chromosome", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
