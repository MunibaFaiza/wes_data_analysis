```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the sample data
data = pd.read_csv("sample_data/gistic_deletion_data.csv")
print(data.head)
```

```python
# Creating Gistic deletion plot
plt.figure(figsize=(10, 6))
sns.heatmap(data.set_index("chromosome"), cmap="Set3", annot=True, fmt=".2f", linewidths=0.5)

plt.title("GISTIC Deletion Plot")
plt.xlabel("Sample")
plt.ylabel("Chromosome")
plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.show()
```
